```python
import matplotlib.pyplot as plt
import matplotlib as mpl
import numpy as np

# 文字颜色 及 支持中文字体
color = 'c'
mpl.rcParams['text.color'] = color
mpl.rcParams['axes.labelcolor'] = color
mpl.rcParams['xtick.color'] = color
mpl.rcParams['ytick.color'] = color

zhfont1 = mpl.font_manager.FontProperties(
    fname="SourceHanSansSC-Bold.otf")
```

```python
x = np.arange(1,11)
y = 2*x +5
fig, ax1 = plt.subplots(1, figsize=(10,10))

ax1.set_title('菜鸟教程 - 测试', fontproperties = zhfont1)
ax1.set_xlabel('x轴', fontproperties = zhfont1)
ax1.set_ylabel('y轴', fontproperties = zhfont1)
ax1.plot(x,y, color = 'lightblue')
plt.show()



```
